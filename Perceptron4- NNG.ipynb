```python
#Perceptron 1
inputs = [1, 2, 3, 4]
targets = [2, 4, 6, 8]

w = 0.1
learning_rate = 0.1

def predict(i):
    return w * i

# Train the network
for _ in range(100):
    pred = [predict(i) for i in inputs]
    errors = [t - p for p, t in zip(pred, targets)]
    cost = sum(errors) / len(targets)

    #print("Targets: ", targets)
    #print("Predictions: ", pred)
    #print("Errors: ", errors)
    print(f"Weight: {w:10f}, Cost: {cost:.6f}")

    w += learning_rate * cost
```

```python
#Perceptron 2
inputs = [1, 2, 3, 4]
targets = [2, 4, 6, 8]

w = 0.1
epochs = 30
learning_rate = 0.1

def predict(x):
    return w * x

# Train the network
for _ in range(epochs):
    pred = [predict(i) for i in inputs]
    errors = [(p - t)**2 for p, t in zip(pred, targets)]  # Cost will generate overflow
    cost = sum(errors) / len(targets)

    # ===================== Derivative ========================
    # Cost using the error derivative
    errors_d = [2 * (p - t) for p, t in zip(pred, targets)]
    weight_d = [e * i for e, i in zip(errors_d, inputs)]
    w -= learning_rate * sum(weight_d) / len(weight_d)

    #print(f"Targets:     {targets}")
    #print(f"Predictions: {pred}")
    #print(f"Errors:      {errors}")
    print(f"Weight (w = {w:.2f}), Cost: {cost:.2f}")
```

```python
#Perceptron 3
inputs = [1, 2, 3, 4]
targets = [4, 6, 8, 10]

w = 0.1
b = 0.3
epochs = 2000

learning_rate = 0.01 # Reduced from 0.1

def predict(x):
    return w * x + b

# Train the network
for _ in range(epochs):
    pred = [predict(i) for i in inputs]
    # The error calculation will no longer overflow with a smaller learning rate
    errors = [(p - t)**2 for p, t in zip(pred, targets)]
    cost = sum(errors) / len(targets)

    # Cost using the error derivative
    errors_d = [2 * (p - t) for p, t in zip(pred, targets)]
    weight_d = [e * i for e, i in zip(errors_d, inputs)]
    # The bias derivative was incorrectly using `errors`. It should use `errors_d`
    bias_d = [e * 1 for e in errors_d] # Corrected the bias derivative calculation

    w -= learning_rate * sum(weight_d) / len(weight_d)
    b -= learning_rate * sum(bias_d) / len(bias_d)

    # Uncomment to debug
    # print(f"Targets: {targets}")
    # print(f"Predictions: {pred}")
    # print(f"Errors: {errors}")
    print(f"Weight: {w:.2f}, Bias: {b:.2f}, Cost: {cost:.2f}")
```

```python
# Entrenamiento de un perceptrón lineal simple

# Datos de entrada y sus objetivos
inputs = [1, 2, 3, 4]
targets = [2, 4, 6, 8]

# Inicialización del peso y tasa de aprendizaje
w = 0.1
learning_rate = 0.1

# Función de predicción
def predict(i):
    return w * i

# Entrenamiento del perceptrón
for _ in range(100):
    pred = [predict(i) for i in inputs]
    errors = [t - p for t, p in zip(targets, pred)]
    cost = sum([e**2 for e in errors]) / len(targets)
    w += learning_rate * sum(errors) / len(targets)
    print(f"Weight: {w:.6f}, Cost: {cost:.6f}")
```
